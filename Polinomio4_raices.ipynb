# Polinomio de grado 4.

## Una raíz.

Dado el polinomio:

\begin{equation}
  p(x) = x^4
\end{equation}  

con raíz $\color{blue}{x^* = 0}$ de multiplicidad 4. Tenemos su gráfica:

```python
import numpy as np
import math
from sympy import *
import matplotlib.pyplot as plt
# --------------------------------------
# Polinomio
x = symbols('x')
p1 = x**4
# --------------------------------------
# Partición
n = 100 # Tamaño de la partición.
# Intervalo a graficar [a,b].
a,b = -5,5
h = (b - a)/n # Incremento de la partición.
xg,yg = [],[]
# --------------------------------------
# Polinomio
for k in range(n):
  xg.append(a+h*k)
  yg.append(p1.evalf(subs={x:xg[k]}))

# --------------------------------------
# Gráfica
plt.plot(xg,yg,'m')
plt.plot(0,0,'b*') #Raíz
# Ejes
plt.axhline(y=0, color='k',linewidth=0.4)
plt.axvline(x=0, color='k',linewidth=0.4)
plt.xlabel("Raíz de $p(x)$.")
# --------------------------------------
```

### Iteración de Newton.

Para el polinomio $P(x) = x^4$, la iteración de Newton, está dada como:
$$
  N(x) = x - \frac{x^4}{4x^3}
$$
simplificando
$$
  N(x) = \frac{3}{4}x
$$
donde la raíz de $p$, $\color{blue}{x^* = 0}$ es punto fijo de $N(x)$, es decir,
$$
  N(\color{blue}{0}) = \color{blue}{0}
$$

La gráfica de $N(x)$ está dada por:








```python
# Función (iteración de Newton para p)
N1 = (3/4)*x
# Se inicializan los vectores para graficar.
yg = []
# --------------------------------------
# Funcíon N(x)
for k in range(n):
  yg.append(N1.evalf(subs={x:xg[k]}))

# --------------------------------------
# Gráfica
plt.plot(xg,yg,'m')
plt.plot(0,0,'b*') #Raíz
# Identidad
plt.plot(xg,xg,'k',linestyle='--',label='Identidad')
# Ejes
plt.axhline(y=0, color='k',linewidth=0.4)
plt.axvline(x=0, color='k',linewidth=0.4)
plt.xlabel("Punto Fijo de $N(x)$.")
# --------------------------------------


```

La identidad está representada por la linea punteada.


Ahora veamos la sucesión de puntos que genera el método de Newton.
$$
  x_{k+1} = x_{k} - \frac{f(x_k)}{f'(x_k)}
$$


```python
# Derifada p'(x) = 4x^3
d_p1 = 4*x**3
# Punto inicial x0
x0  = 3       # punto inicial
x1  = x0+1    # iniciamos segunda iteración
xk  = []      # vector donde se guardan las iteraciones
yk  = []
xk.append(x0)
yk.append(0)
tol = 10**(-8)# tolerancia del método
while abs(x0 - x1) > tol:
  x1  = x0 - (p1.evalf(subs={x:x0}))/(d_p1.evalf(subs={x:x0}))
  xk.append(x1)
  yk.append(0) # vector auxiliar para graficar los puntos x_k
  aux = x1
  x1  = x0
  x0  = aux

nk = len(xk)  # Número de iteraciones.


plt.plot(0,0,'b*',linewidth=4)
#plt.plot(xk,yk,'b|')
plt.axhline(y=0, color='k',linewidth=0.4)

for i in range(nk):
  plt.plot(xk[i],yk[i],'b|')
  plt.pause(0.1)




```
